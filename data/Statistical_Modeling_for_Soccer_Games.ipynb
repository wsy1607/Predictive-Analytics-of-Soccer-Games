### Assumption

In soccer games, we assume the number of goals scored by each team are Poisson Distributions.

$$
G_{ij} = X_{ij} + h,
$$
$$
X_{ij} \sim Poisson(\lambda_{ij}),
$$
$$
\lambda_{ij} = a_i - d_j + \epsilon_{ij}.
$$

The basis of our model on predicting the result of soccer game is the idea that the number of goals scored by one team in a game follows Poisson distribution. If we divide a game into infinitely many time intervals, in each interval the probability that a goal is scored by one team (say team A) is constant, if we ignore the effect of substitution and sent-offs. This probability is only dependent on the attacking power of team A, the defending power of team B and the home team advantage. So in each time interval the number of goals scored follows binomial distribution with constant probability. The total number of goals scored equals the sum of the number of goals scored in all the time intervals. So it is the sum of infinite binomial distribution, which is Poisson distribution. 
The exact number of goals scored by team A is determined by the Poisson parameter $\lambda$. As we mentioned before, it is related with the attacking power of team A, the defending power of team B. Here we consider the simplest case, which is the attacking factor of team A minus the defending power of team B. We use a team’s past record of goals for and goals against to perform maximum likelihood analysis to estimate team’s attacking and defending power. We also assumed that the attacking and defending power can be related to game statistics of teams using OLS analysis. Now, as long as we have the game statistics of two teams, we can plug them into the OLS model to estimate their attacking and defending power. Then we plug these powers in to the calculation of the Poisson $\lambda$ to estimate the result of the game between them. Finally we add the home team advantage (which will be discussed later).

However, the analysis above has a major problem, which is that in those six club leagues in Europe, including England, Germany, Italy, France, Spain and Dutch, the rules are slightly different from the rules in the World Cup tournament. In club leagues, winning a game will get 3 points, and get 1 for a tired game and 0 for losing a game. So it encourages teams to score more goals. In the World Cup group stage, it is the same. But when we step into the knockout (second) stage, it is a single-elimination tournament in which teams play each other in one-off matches, with extra time and penalty shootouts used to decide the winner if necessary. So it becomes a fair game for teams with better attacking ability and other teams with better defensing ability.

Then we need to do the same analysis for some tournaments with the same rules as in the World Cup to make it more comparable. ## Regression model for prediction number of goals scored

By our assumption, the number of goals scored in a single game follows a Poisson distribution with parameter $\lambda$. Firstly, we want to estimate $\hat{\lambda}$ by MLE, which is just the mean of goals scored in a series of games. Then we run OLS for $\hat{\lambda}$ as dependent variable and other game statistics as the design matrix. Here, our traing data is from UEFA data sets through resent 10 years. The model will be applied in our predicting process for World Cup 2014.

# Players Performance Analysus

We would like to see what is the relation between players' statistics and their performance. We use rating to measure a player's performance. The rating is given by authority groups which consist of commentators, newspapers and broadcasts. We assume the relationship between players statistics and their rating for each single game is linear.

## Linear Model

First of all, we use a wider range of variables.

```python
#in this script, we are going to run linear regression on players' statistics to see the relationship statistics related to 
#the html source code for different year has the same pattern, so we can read data at once
#read players data
filename = ['premierplayers0910.txt','premierplayers1011.txt','premierplayers1112.txt','premierplayers1213.txt','premierplayers1314.txt','laligaplayers0910.txt','laligaplayers1011.txt',
'laligaplayers1112.txt','laligaplayers1213.txt','serieAplayers0910.txt','serieAplayers1011.txt','serieAplayers1112.txt','serieAplayers1213.txt',
'bundesligaplayers0910.txt','bundesligaplayers1011.txt','bundesligaplayers1112.txt','bundesligaplayers1213.txt','frenchplayers0910.txt','frenchplayers1011.txt',
'frenchplayers1112.txt','frenchplayers1213.txt','brazilplayers13.txt','championplayers1314.txt','dutchplayers1314.txt','majorplayers13.txt',
'russianplayers1314.txt','laligaplayers1314.txt','serieAplayers1314.txt','bundesligaplayers1314.txt','frenchplayers1314.txt']

columnnames = ['TeamId','PlayerId','Field','GameStarted','SubOn','SubOff','Yellow','SecondYellow','Red','Goals','Assists','TotalPasses','AccuratePasses',
'AerialWon','AerialLost','Rating','ManOfTheMatch','TotalTackles','Interceptions','Fouls','OffsidesWon','TotalClearances','WasDribbled','TotalShots',
'ShotsOnTarget','ShotsBlocked','OwnGoals','KeyPasses','Dribbles','WasFouled','Offsides','Dispossesed','Turnovers','TotalCrosses','AccurateCrosses',
'TotalLongBalls','AccurateLongBalls','TotalThroughBalls','AccurateThroughBalls','Height','Weight','Ranking','Age']
data = []
for name in filename:
	txt = open(name)
	data1 = txt.read()
	data2 = data1.split('DataStore.prime')
	data3 = data2[1].split('}]]);')[0]
	data4 = data3.split('[{')[1]
	regex = re.compile(':[0-9]+\.*[0-9]*')
	data5 = regex.findall(data4)
	data6 = ' '+' '.join(data5)
	data7 = data6.split(' :')[1:]
	data.append(data7)

data8 = np.array(data).reshape(600,43)
data9 = DataFrame(data8,columns = columnnames,dtype='float')
data9['const'] = np.ones(600)
#reorder the index
ind = np.random.permutation(range(0,600))
mod1 = sm.OLS(data9.ix[0:600,15], data9.ix[0:600,[5,6,8,9,10,11,13,17,18,19,21,24,26,27,28,29,30,32,39,40,42,43]]).fit()
#Inspect the results
print mod1.summary()
```

From the results, the $R^2$ is large but we have strong multicolinearity and many of these predictors are travail and should be taken off.

```python
mod2 = sm.OLS(data9.ix[0:600,15], data9.ix[0:600,[5,9,10,13,17,19,27,28,30,32,43]]).fit()
#Inspect the results
print mod2.summary()
```

From the regression table, we can see that the results have been improved and all the predictors pass the t-test with a 0.05 significant level. When a player get substituted off in the game, get fouls, get off side calls or turnovers, he would lose some points for his rating. On the other hand, goals, assists, tackles, aerial won, key passes, dribbles will benefit the player for a better rating. Among all those players’ statistics, the most important one is goal scored which is obviously. For the constant, it is interesting to discover that if a player does nothing in the court, he will receive probably a rating about 7.22 which is not bad. But it is unrealistic.

```python
predictions = mod2.predict(data9.ix[0:600,[5,9,10,13,17,19,27,28,30,32,43]])
fig = plt.figure()
ax1 = fig.add_subplot(2, 1, 1)
ax1.plot(data9.ix[ind,15], color = 'k', label = 'true', marker='o')
ax1.plot(predictions[ind], color = 'r', label = 'model',linestyle='dashed', marker='o')
ax1.hlines(8.5,0,600, color = 'b')
plt.ylim([6.5, 10])
plt.xlim([0, 600])
ax1.legend(loc='best')
ticks = ax1.set_xticks([0,100,200,300,400,500,600])
ax1.set_ylabel('rating')
ax1.set_title('players performance regression model')
ax2 = fig.add_subplot(2, 1, 2)
ax2.hist(data9.ix[ind,15]-predictions)
ax2.set_ylabel('frequencies')
ax2.set_title('histogram of prediction errors')
```

This graph summarizes our result of player performance rating analysis. Here black dots represented performance rating given by media or soccer experts, and red dots represent actual performance estimated from players’ game statistics. We see that in general the pairs of black and red dots are quite close to each other except for a few pairs. This means that media and soccer experts are, in most cases reliable. We also noticed that for pairs that the black and red dots are quite far way, the black dots are always higher than the red dots. This indicates that in those cases when the experts and media are not reliable, they usually tend to exaggerate play’s performance rather than underestimating them. We also noticed that these cases usually happen on players with quite high rating. This can warn those managers to notice that performance rating sometimes can be misleading by exaggerating the performance of top tier players. 

However, there still exits some multicorlinearity

## Cross Validation

Since there should be some outliers, we take the leave-one-out cross validation method to do a model selection.

```python
val = []
for i in ind:
	mod = sm.OLS(data9[data9.index != i].ix[:,15], data9[data9.index != i].ix[:,[5,9,10,13,17,19,27,28,30,32,43]]).fit()
	prediction = mod.predict(data9.ix[i,[5,9,10,13,17,19,27,28,30,32,43]])
	val.append((prediction-data9.ix[i,15])**2)

#the one with minimum value is "ind[val.index(min(val))]" 
mod = sm.OLS(data9[data9.index != ind[val.index(min(val))]].ix[:,15], data9[data9.index != ind[val.index(min(val))]].ix[:,[5,9,10,13,17,19,27,28,30,32,43]]).fit()
print mod.summary()
```

The result shows that it is a little bit better than the previous one. Then we make the same plot again. 

```python
prediction = mod.predict(data9.ix[:,[5,9,10,13,17,19,27,28,30,32,43]])
fig = plt.figure()
ax1 = fig.add_subplot(2, 1, 1)
ax1.plot(data9.ix[ind,15], color = 'k', label = 'true', marker='o',alpha = 0.5)
ax1.plot(prediction[ind], color = 'r', label = 'model',linestyle='dashed', marker='o',alpha = 0.8)
ax1.hlines(8.5,0,600, color = 'b')
plt.ylim([6.5, 10])
plt.xlim([0, 600])
ax1.legend(loc='best')
ticks = ax1.set_xticks([0,100,200,300,400,500,600])
ax1.set_ylabel('rating')
ax1.set_title('cross validation model')
ax2 = fig.add_subplot(2, 1, 2)
ax2.hist(data9.ix[ind,15]-prediction)
ax2.set_ylabel('frequencies')
ax2.set_title('histogram of prediction errors')
```
